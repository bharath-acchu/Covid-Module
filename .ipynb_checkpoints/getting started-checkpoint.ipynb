## How to use John Hopkins University API



```python
# country wise details

from covid import Covid

covid = Covid()
covid.get_data()
```

#### You can get status by country wise

This will be helpful when you nedd to know the avaliable names of countries.By using `covid.list_countries()` we get to know the country exact name that can be used.

```python
countries = covid.list_countries()
```

```python
countries
```

```python
india_cases = covid.get_status_by_country_name("india")
```

```python
# status of India

india_cases
```

## You can access using ID 

```python
India_cases = covid.get_status_by_country_id(91)
```

```python
India_cases
```

## Get total active cases

```python
active = covid.get_total_active_cases()
```

```python
active
```

## Get Total confirmed cases


```python
confirmed = covid.get_total_confirmed_cases()
```

```python
confirmed
```

## Get total recovered cases

```python
recovered = covid.get_total_recovered()
```

```python
recovered
```

## Get Total Deaths

```python
deaths = covid.get_total_deaths()
```

```python
deaths
```

## How to use worldmeters.info

```python
covid1 = Covid(source="worldometers")
```

```python
# check the source
covid1.source
```

```python
covid1.get_data()
```

```python
covid1.get_status_by_country_name("china")
```

```python
covid1.list_countries()
```

## Getting total active cases, total confirmed cases , total recovered cases , total deaths are same as above mentioned syntax.

# Using CLI

* ./cli.py covid

or 


* ./cli.py covid -s john_hopkins


### In above syntax john_hopkins can be replaced with worldometers for worldometer source

## List countries 

./cli.py covid -s worldometers --list-countries

## Get Staus By country name 

./cli.py covid -s worldometers -c sweden

## Get total Active cases

./cli.py covid -s worldometers -o active


## Get Total confirmed cases

./cli.py covid -s worldometers -o confirmed



## Get total Recovered cases

./cli.py covid -s worldometers -o recovered

## Get total Deaths

./cli.py covid -s worldometers -o deaths
